# Numerical Python

## Basics

```python
import numpy as np
```

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

x+y # notice it doesn't concaetenate
```

```python
# Matrixes - Multidimensional Arrays
x = np.array([
            [1, 2, 3],
            [4, 5, 6],
                    ])
x
x.ndim # Number Dimensions
x.dtype # Composed of int32's
x.sum() # np.sum(x)
x.size  # Number elements
x.shape
x.reshape( (3, 2) ) # 3 elements 2nd Dim, 2 elements first Dimension
x.T # Transpose - Rows exchange with columns
```

```python
print(np.sqrt(x)) # Square Root Each Element
print(np.mean(x)) # Mean of all 
print(np.var(x)) # Variance of all
print(np.std(x)) # Standard Deviation of all
x**2 # each element squared
```

## Random

```python
# Accepts Keyword Arguments i.e., hello(first=1, second=4)
# Some restrict to positional arguments i.e., hello(1, 2)
y = np.random.normal(size=25)
x = y + np.random.normal(loc=50, scale=1, size=25)

# Correlation Matrix
np.corrcoef(x, y)
```

```python
# If we set custom SEED, we will get same random numbers each time for that SEED
rng = np.random.default_rng(1303)
rng2 = np.random.default_rng(1303);

# rng has all properties of random
print(rng.normal(size=2))
print(rng2.normal(size=2))
```

```python
# Mean across axis
np.mean(x, axis=0)
```

# Graphics

```python
from matplotlib.pyplot import subplots

x=rng.standard_normal(size=100)
y=rng.standard_normal(size=100)
```

```python
figure, axes = subplots( figsize=(10,10), nrows=2, ncols=2)

# Axes Properties
axes[0,0].plot(x, y) # Line Plot
axes[0,0].set_title("Relation b/w Number of Rooms and House Price")
axes[0,0].set_xlabel("Number of Rooms")
axes[0,0].set_ylabel("House Price ($)")
axes[0,0].set_xlim( (1,-1) )

axes[0,1].plot(x, y, 'yo') # Scatter Plot
axes[1,0].scatter(x, y, marker='o'); # SemiColon to prevent it from printing text, remove to see the text

# Figure Properties
figure.set_size_inches(10, 10) # Resize
figure.savefig("Figure.png", dpi=400)
```

```python
x = np.linspace(-np.pi, np.pi, 50) # 50 numbers from -3.14, till 3.14 with equal gaps b/w
y = x
f = np.multiply.outer(np.cos(x), 1 / (1 + x**2))

fig, axes = subplots(figsize=(8,8))
axes.contour(x, x, f, levels=75);
axes.imshow(f); # Heatmap (same as contour except colo-coded plot where colors depend on Z value)
```
